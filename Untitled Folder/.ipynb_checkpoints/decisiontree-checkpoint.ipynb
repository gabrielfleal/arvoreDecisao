```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
from sklearn import preprocessing
import graphviz

#http://dataaspirant.com/2017/02/01/decision-tree-algorithm-python-with-scikit-learn/
```

```python
clf = DecisionTreeClassifier(random_state=0)
dataset = pd.read_csv("survey_formatado_country_sem_colunas.csv", delimiter=";")
dataset_array = dataset.as_matrix()
original_headers = list(dataset.columns.values)
dataset.shape


le = preprocessing.LabelEncoder()

x= dataset.values[:,0:22]
y= dataset.values[:,23]
cont = 22

for c in range(cont):
    x[:,c]=le.fit_transform(x[:,c].astype(str))
    
y= le.fit_transform(y)
clf = clf.fit(x,y)
cv_scores = cross_val_score(clf, x, y, cv=10)
```

```python
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size = 0.3, random_state = 100)

#utilizando indice de gini
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,
                               max_depth=3, min_samples_leaf=5)
clf_gini.fit(X_train, y_train)
```

```python
##utilizando ganho de informação
clf_entropy = DecisionTreeClassifier(criterion = "entropy", random_state = 100,
 max_depth=3, min_samples_leaf=5)
clf_entropy.fit(X_train, y_train)
```

```python
classe_names = clf_gini.classes_
feature_names = dataset.columns[0:22]
#grafico do gini
dot_data = tree.export_graphviz(clf_gini, out_file=None, feature_names=feature_names,class_names=["No", "Yes"] ) 
graph = graphviz.Source(dot_data)
graph
```
